```python
def for_loop(n):
    total_sum = 0
    
    for i in range(n):
        total_sum += i**2
    
    return total_sum
```

```python
def while_loop(n):
    total_sum = 0
    i = 0
    
    while i <= n:
        total_sum += i**2
        i += 1
        
    return total_sum
```

```python
def maths_loop(n):
    return (n)*(n+1)*(2*n+1)/6
```

```python
entrada = 10000000
%timeit for_loop(entrada)
%timeit while_loop(entrada)
%timeit maths_loop(entrada)
```
